```python
import ipywidgets as widgets
from IPython.display import clear_output
import random
import numpy as np


#WARNING READING MY CODE MAY BE DETRIMENTAL TO YOUR MENTAL HEALTH.





def sixes_probability(x, y): #Calculates the probability of rolling x or more sixes on y dice. Used to calculate odds of sucess and odds of a mishap.


    #Create a 2d array to store the probabilities for each combination of rolls.
    probabilities = np.zeros((y + 1, x + 1))

    #Starting assumptions, No dice rolled or all sixes rolled.
    probabilities[0, 0] = 1
    probabilities[y, x] = 1

    #Calculate probabilities for each number of dice rolled.
    for i in range(1, y + 1):
        for j in range(x + 1):
            # Probability of rolling a six is 1/6, not rolling a six is 5/6.
            probabilities[i, j] = (5 / 6) * probabilities[i - 1, j] + (1 / 6) * probabilities[i - 1, max(j - 1, 0)]

    # Return the probability.
    return probabilities[y, x]




#Paint the UI. Create and define the inputs and paint them in a box.
Diceused = widgets.RadioButtons(options=[str(i) for i in range(1, 11)])
SpellTarget = widgets.RadioButtons(options=[str(i) for i in range(1, 11)])
button = widgets.Button(description='Cast')
horizontal_box = widgets.HBox([Diceused, SpellTarget, button])
print("To use this program put the number of spell dice you are using into the left collum and level of the spell you are attempting to cast in the right. Click cast and the computer will tell you the results.")
display(horizontal_box)

#The button that does the real work. Do this so it doesn't try and recalculate every time you change a value.

def on_button_clicked(b):
  clear_output(wait=True)#Clear and redraw the display so it does not keep scrolling.
  display(horizontal_box)#Remove these lines for persistance.

  #index = 0

  #Create an array and fill it with random numbers between 1-6 utill we have all the used spell dice accounted for.
  diceindex = []
  for i in range(int(Diceused.value)):
    diceindex.append(random.randint(1,6))
  print (diceindex)

  #The spell power starts as the number of spell dice, each six rolled boosts this by one. If the spell power is more then the required power of the spell, the spell works. If it's less the spell fails.
  Spellpower = (int(Diceused.value)) + diceindex.count(6)
  if Spellpower >= int(SpellTarget.value):
    print ("The spell is cast with a power of ",Spellpower)
  else:
    print ("The spell failed")

  #A one or two will result in losing the die, a three, four, five or six will return to the dice pool. We inform the player of the results.
  lostdice = diceindex.count(1) + diceindex.count(2)
  print (f"You used {int(Diceused.value)} Spelldice. {lostdice} of these dice were lost. {int(Diceused.value) - lostdice} Dice return to your pool")
  #Rolling two or more ones will result in a mishap that gets worse with the number of ones rolled.
  if diceindex.count(1) > 1:
    print (f"There was a level {diceindex.count(1)-1} mishap.")

  #Now we tell the player the odds.

  if int(Diceused.value) < int(SpellTarget.value):#This only matters if the caster used fewer dice then the spell required. Here we tell them the odds that the spell would sucessed anyway.
    SucsessOdds = (sixes_probability(int(SpellTarget.value)-int(Diceused.value), int(Diceused.value))) * 100
    print (f"The odds of sucess were {SucsessOdds:.2f}%")

  if int(Diceused.value) > 1:#If the caster used more then 1 die then there is a chance of a mishap. Reuses the function as above for two or more doubles.
    MishapOdds = (sixes_probability(2, int(Diceused.value))) * 100
    print (f"The odds of a mishap were {MishapOdds:.2f}%")


button.on_click(on_button_clicked) #(maine)







```
